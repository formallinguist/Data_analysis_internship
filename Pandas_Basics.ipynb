# Pandas Series
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.



```python
import pandas as pd
```

```python
# Series Method we see series is a 1 D array.

series = pd.Series([1, 7, 2, 8, 5]) #Default index

print(series)
```

```python
# Series Method with Index

series = pd.Series([1, 7, 2, 8, 5], index = ["a", "b", "c","d","e"]) #assigning the as per the choice

print(series)
```

```python
# Accessing Series elements
print(series[2])
```

```python
print(series["c"])
```

# Pandas DataFrame : Multiple rows and multiple columns, series has single row

A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns.

```python
# DataFrame Method on list
data = pd.DataFrame([1,2,3,4,5])
print(data)
```

```python
# DataFrame using Dictionary

data = pd.DataFrame({'Name':['Tom', 'Jack', 'Steve', 'Ricky','Sam','Will','Root'],
                     'Age':[28,34,29,42,17,17,45]})
data.head(10)
```

```python
data.head(5)
```

```python
# DataFrame with Index

data = pd.DataFrame({'Name':['Tom', 'Jack', 'Steve', 'Ricky','Sam','Will','Root'],
                     'Age':[28,34,29,42,17,17,45]}, index=['C1','C2','C3','C4','C5','C6','C7'])
data.head(10)
```

#Column selection

```python
data['Name'] #Acessing data for sorting purpose.
```

```python
data['Age']
```

```python
data.head(10)
```

# Row Selection

```python
data.loc["C1","Age"] #In loc we use string values
```

```python
data.iloc[3,0]#In iloc we take numerical values as they are integer values.
```

# Pandas Concatenation

It is used to concatenate pandas objects along a particular axis with optional set logic along the other axes.

```python
names = pd.DataFrame({'Name':['Tom', 'Jack', 'Steve', 'Ricky','Sam','Will','Root']})
ages = pd.DataFrame({'Age':[28,34,29,42,17,17,45]})
```

```python
names2 = pd.DataFrame({'Name':['Joe', 'Pope', 'Buttler', 'Ben','James','Cook']})
```

```python
names2
```

```python
ages
```

```python
pd.concat([names, ages], axis = 1)                     # Concatenation horizontally
```

```python
pd.concat([names, names2])                # Concatenation vertically
```

```python
pd.concat([names, names2],axis=0)
```

```python
pd.concat([names, names2]).reset_index()
```
